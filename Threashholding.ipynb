```python
import cv2 as cv
from cv2 import imread
import os
from os.path import join
import matplotlib.pyplot as plt
from  matplotlib.pyplot import imshow


base_path = '/Users/apple/Downloads'
images = os.listdir(join(base_path, 'images'))
img = imread(join(base_path, 'images', images[0]), 0)

# ADAPTIVE_THRESH_GAUSSIAN_C
# THRESH_BINARY
# ADAPTIVE_THRESH_GAUSSIAN_C

th3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY,401,50)

imshow(th3, cmap='gray')
```

```python
blur = cv2.GaussianBlur(img,(441,441),200, 200)

edges = cv2.Canny(blur,0,2)
plt.gcf().set_size_inches(10, 10)
imshow(edges, cmap='gray')
```
